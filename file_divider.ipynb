```python
import os
import pandas as pd
from collections import OrderedDict
import ast
import math
```

```python
file_name = '1682254038522535'
```

```python
def convert_to_float_list(string_list):
    nested_list = ast.literal_eval(string_list)
    float_list = [[float(item) for item in sublist] for sublist in nested_list]
    return float_list
```

```python
data = pd.read_csv(f'sensors_labeling_data/klata/CLIMBING/{file_name}.csv',sep=';', header=None, names=['Accelerometer', 'Gyroscope', 'Magnetometer','Duration'])
data['Accelerometer'] = data['Accelerometer'].apply(convert_to_float_list)
data['Gyroscope'] = data['Gyroscope'].apply(convert_to_float_list)
data['Magnetometer'] = data['Magnetometer'].apply(convert_to_float_list)
data
```

```python
data['Accelerometer'][0]
```

```python
half_length_acc = math.floor(len(data['Accelerometer'][0]) / 2)
half_length_gyro = math.floor(len(data['Gyroscope'][0]) / 2)
half_length_mag = math.floor(len(data['Magnetometer'][0]) / 2)

df1 = data.copy()
df2 = data.copy()

df1['Accelerometer'][0] = data['Accelerometer'][0][:half_length_acc]
df2['Accelerometer'][0] = data['Accelerometer'][0][half_length_acc:]

df1['Gyroscope'][0] = data['Gyroscope'][0][:half_length_gyro]
df2['Gyroscope'][0] = data['Gyroscope'][0][half_length_gyro:]

df1['Magnetometer'][0] = data['Magnetometer'][0][:half_length_mag]
df2['Magnetometer'][0] = data['Magnetometer'][0][half_length_mag:]

# Dzielimy wartość w kolumnie 'Duration'
df1['Duration'] = math.floor(data['Duration'][0] / 2)
df2['Duration'] = math.floor(data['Duration'][0] / 2)
```
